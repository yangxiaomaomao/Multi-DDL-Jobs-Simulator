```python
import csv
import sys
import os
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
import numpy as np
iter_time_list = {
    "ideal": np.array([4,10,3.75]),
    "Crux":  np.array([4.18,37.8,3.87]),
    "Crux+": np.array([4.19,33.6,3.96]),
    "worst": np.array([17.9,10.5,4.21]),
    # "equal": np.array([7.18, 14.6, 4.12]),
    "Mine":   np.array([7.95,15.7,3.9])
}
res_dict = {}
for method, times in iter_time_list.items():
    if method != "ideal":
        res_dict[method] = np.sum(times / iter_time_list["ideal"])

predict = {
    # "ideal": 3,
    "Crux":  5.6,
    "Crux+": 5.42,
    "worst": 6.07,
    "Mine":   4.7
}

# 方法名
methods = list(res_dict.keys())

# 数据值
res_values = np.array(list(res_dict.values()))
predict_values = np.array(list(predict.values()))

# 设置柱子的宽度
bar_width = 0.35

# 计算 x 轴位置
x = np.arange(len(methods))

# 创建柱状图
fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(x - bar_width/2, res_values, bar_width, label='Simulator', color='#2b4652', alpha=0.7)
ax.bar(x + bar_width/2, predict_values, bar_width, label='Predict', color='#dd7055', alpha=0.7)

# 设置 x 轴标签
ax.set_xticks(x)
ax.set_xticklabels(methods, fontsize=20)

# 添加标题和标签
ax.set_ylabel('Value', fontsize=20)
ax.set_title('Comparison of Methods: Simulator vs Predict', fontsize=20)
ax.legend(fontsize=15)

# 显示网格
ax.yaxis.grid(True, linestyle='--', alpha=0.7)

plt.savefig("simulator_vs_predict.pdf", bbox_inches='tight')
```
